```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
ssom1=pd.read_excel("./썸트렌드_언급량_210901.xlsx")
ssom2=pd.read_excel("./썸트렌드_언급량_220901.xlsx")
ssom3=pd.read_excel("./썸트렌드_언급량_230901.xlsx")
```

```python
ssom=pd.concat([ssom1,ssom2,ssom3])
weekly=np.arange(len(ssom['날짜']))

plt.plot(weekly,ssom['합계'])

plt.title('워케이션 관심도 추세')
plt.xlabel("기간")
plt.ylabel('언급량')
```
