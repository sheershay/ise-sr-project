```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
pub=pd.read_excel("C://Users/user/Documents/code/종설/NewsResult.xlsx")
per1=pd.read_excel("C:/Users/user/Documents/code/종설/썸트렌드_연관어_210901.xlsx")
per2=pd.read_excel("C:/Users/user/Documents/code/종설/썸트렌드_연관어_220901.xlsx")
per3=pd.read_excel("C:/Users/user/Documents/code/종설/썸트렌드_연관어_230901.xlsx")
```

```python
per_v1=pd.concat([per1,per2,per3])
#per_v2=pd.concat([per1,per2,per3],axis=1)

per_v1=per_v1.drop(columns=['순위'])
per_v1.sort_values(by=['건수'],ascending=False,inplace=True)

per_v1[per_v1.duplicated(['연관어'])==True] #669 단어 반복해서 나타남 #원래 1500행

samewords=per_v1.groupby('연관어').sum()
samewords.sort_values(by=['건수'],inplace=True,ascending=False)
samewords
```

```python
x=np.arange(len(samewords))
per_num=samewords['건수']
plt.scatter(x,per_num)

threshold=1500

above_threshold = per_num >= threshold
plt.scatter(x[above_threshold], per_num[above_threshold], color='orange') 
plt.ylabel('Frequency')
```
