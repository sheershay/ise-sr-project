```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
per1=pd.read_excel("./썸트렌드_연관어_210901.xlsx")
per2=pd.read_excel("./썸트렌드_연관어_220901.xlsx")
per3=pd.read_excel("./썸트렌드_연관어_230901.xlsx")
```

```python
per_v1=pd.concat([per1,per2,per3])
per_v1.drop(columns=['순위'],inplace=True)
per_v1.drop_duplicates(subset=['연관어'],inplace=True)
```

```python
per_v1.sort_values(by=['건수'],axis=0,ascending=False,inplace=True)
per_v1
```

```python
x=np.arange(len(per_v1))
per_num=per_v1['건수']
plt.scatter(x,per_num,s=100,alpha=0.3)

threshold=500

above_threshold = per_num >= threshold
plt.scatter(x[above_threshold], per_num[above_threshold], color='red') 

plt.xlabel('연관어')
plt.ylabel('검색량')
plt.title('워케이션 연관어 분포도')
```

```python
per_region=per_v1.loc[per_v1['카테고리 소분류']=='지역/자연']
per_region['연관어'].is_unique
per_region.sort_values(by=['건수'],axis=0,ascending=False)
per_region
```

```python
x2=np.arange(len(per_region))
y2=per_region['건수']

plt.scatter(x2,y2,alpha=0.3)

plt.scatter(x2[0:6], y2[0:6], color='red') 

plt.title('"지역" 연관어 분포도')
plt.xlabel('지역 연관어')
plt.ylabel('검색량')
```
